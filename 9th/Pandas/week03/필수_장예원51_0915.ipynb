- horsepower: 마력
- weight: 무게
- acceleration: 가속력 
- model_year: 모델 연도

- name: 이름

```python
# Derived Variable

mpg_dv = mpg.copy()
```

```python
## 효율 관련 변수

mpg_dv['km_per_liter'] = mpg_dv['mpg'] * 0.425144

mpg_dv['fuel_consumption'] = 100 / mpg_dv['km_per_liter']

mpg_dv['hp_per_displacement'] = mpg_dv['horsepower'] / mpg_dv['displacement']

mpg_dv['efficiency_index'] = mpg_dv['mpg'] / mpg_dv['weight']
```

```python
## 성능 관련 변수

# mpg_dv['hp_per_weight'] = mpg_dv['horsepower'] / mpg_dv['weight']
mpg_dv['power_roweight_ratio'] = mpg_dv['horsepower'] / mpg_dv['weight']

mpg_dv['acceleration_category'] = pd.cut(mpg_dv['acceleration'], bins=[0, 10, 15, 20], labels=['빠름', '보통', '느림'])

mpg_dv['acc_per_weight'] = mpg_dv['acceleration'] / mpg_dv['weight']

mpg_dv['disp_per_cylinder'] = mpg_dv['displacement'] / mpg_dv['cylinders']
```

```python
## 연식 및 연료 효율 관련 변수

year = 2024
mpg_dv['car_age'] = year - mpg_dv['model_year']

mpg_dv['fuel_efficiency_category'] = pd.cut(mpg_dv['mpg'], bins=[0, 15, 25, 50], labels=['저효율', '보통', '고효율'])
```

```python
## 엔진 및 차량 타입 관련 변수

mpg_dv['engine_size_category'] = pd.cut(mpg_dv['displacement'], bins=[0, 200, 400, 600], labels=['소형', '중형', '대형'])

mpg_dv['origin_category'] = mpg_dv['origin'].replace({1: 'USA', 2: 'Europe', 3: 'Japan'})
```
