```python
pc_gr = tt.groupby('pclass')
```

```python
# 어떤 식으로 그룹이 묶였나??

for i, j in pc_gr:
    print(i, j)
```

```python
# groupby를 한 후에 1번으로 값을 넣으면 내가 그룹한 값의 데이터프레임 반환

pc_gr.get_group(2)
```

```python
ps_gr.get_group((2, 'female'))
```

```python
# groupby 후, 특정 통계치를 추출하고 싶은 경우
# sum, mean, median, ,,,
# agg(새로운 컬럼명 = (기준 컬럼, 통계치)) ... n 증가

tt.groupby('pclass').agg(fare_sum=('fare', sum))
```

```python
# na 값 제외한 계산이 된다

tt.groupby('pclass').agg(fare_sum = ('fare', 'sum'),
                         fare_mead = ('fare', 'mean'),
                         age_mean = ('age', 'mean'),
                         age_cnt = ('age', 'count'))
```

## qcut, cut

- **qcut**:
    데이터를 분위수(Quantitle) 따라 구간을 나눈다.
    - 분위수 기준으로 균등하게 데이터를 나누는 함수
    - **pd.qcut(x, q, labels = ?, retbins = ?, precision = ?, duplications = ?)**
        - **x**: 구간 데이터 나눌 시리즈 형태 또는 배열
        - **q**: 구간 나눌 분위수 e.g 4면 4분위수
        - **labels**: 각 구간에 붙일 이름
        - **retbins**: 구간의 경계를 함께 반환 여부
        - **precision**: 구간의 경계의 소숫점 자릿수

        - **duplications**:

- **cut**:
    고정된 값의 구간으로 나누는 함수. 데이터 값 범위에 따라 일정한 크기의 구간을 정의하고, 각 데이터가 해당하는 구간을 속하게 만들면 된다.
    - **pd.cut(x, binx, right=?, 나머지 동일, include_lowest=가장 낮은 구간 포함 여부)**
    - 위와 동일한 하이퍼 파라미터

```python
# qcut 예시 데이터

import numpy as np

data = np.random.randn(10)
print(data)
```

```python
res_qcut = pd.qcut(data, q=4, labels=['1사분위', '2사분위', '3사분위', '4사분위'])
print(res_qcut)
```

```python
# cut 예시 데이터

data = np.random.randn(10) * 100
print(data)
```

```python
res_cut = pd.cut(data, bins=4, labels=['낮음', '중간', '높음', '매우높음'])
print(res_cut)
```

```python
'''
타이타닉 데이터를 예시로 데이터를 나누자!

1. age 기준으로 도메인 나누기
    na 값은 Unknown
    19세 미만 child
    19~70 Adult
    나머지 Senior
    
2. fare 구간
    qcut, cut 나누기
    
구간화를 범주로 만들어 파생변수를 만들고 확인해보자!

데이터프레임에 특정 함수를 전체에 적용시킬 때 aply()
apply(전처리할 함수)

age에 대한 함수 만들기
'''

def age_cat(age):
    if pd.isnull(age):
        return 'Unknown'
    elif age < 19:
        return 'child'
    elif age < 70:
        return 'Adult'
    else:
        return 'Senior'
```

```python
# apply()를 이용해 전처리

tt['age_group'] = tt['age'].apply(age_cat)
```

```python
# are 요금에 대한 전처리

tt['fare_qcut'] = pd.qcut(tt['fare'], q = 4, labels = ['1사분위', '2사분위', '3사분위', '4사분위'])
tt['fare_cut'] = pd.cut(tt['fare'], bins = 4, labels = ['1사분위', '2사분위', '3사분위', '4사분위'])
```

```python
tt.fare_qcut.value_counts()
```

```python
tt.fare_cut.value_counts()
```

## 데이터 병합

- 붙이는 것
    - 왜 붙일까?
    - 기업에서는 데이터를 DB에 테이블로 나눠서 보관
    - ML 전처리 데이터 병합

- merge _병합시간에 배울 예정_
    - 공통된 컬럼을 가지고 병합
    - **join**:
    공통 컬럼 기준을 지정한 후 inner, left, right, outer 조인 연산 방법
    - **merge(df1, df2, how = '조인 연산 방법', on = '공통 컬럼')**
        
- concat
    - 덩어리와 덩어리를 붙이는 것
    - df + df을 그대로 붙이는 것
    - df1, df2
        - 행: 가로로 길어짐
        - 열: 세로로 길어짐

```python
test1 = pd.DataFrame({'학회원 분반 번호': [1, 2, 3, 4, 5],
                      '학회원 분반': ['Python', 'R', 'SQL', '데이터 전처리', '데이터 모델링']})

test2 = pd.DataFrame({'학회원 분반 번호': [1, 2, 3, 4, 6],
                      '학회원 수': [100, 200, 300, 400, 500]})
```

```python
display(test1)
```

```python
display(test2)
```

```python
pd.merge(test1, test2, how='inner', on='학회원 분반 번호')
```

```python
pd.concat([test1, test2], axis = 0)
```

```python
pd.concat([test1, test2], axis = 1)
```
